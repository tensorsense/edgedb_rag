```python
from pathlib import Path
from tqdm import tqdm
import sys

sys.path.append(Path("..").resolve().as_posix())
```

```python
old_path = Path("eval_old_hard_2024-02-23-15-12-26.jsonl")
new_path = Path("eval_new_hard_2024-02-23-09-47-51.jsonl")

assert old_path.exists() and new_path.exists()
```

```python
import pandas as pd

old_df = pd.read_json(old_path, lines=True)
new_df = pd.read_json(new_path, lines=True)

eval_df = pd.merge(left=old_df[["query", "response"]], right=new_df[["query", "response"]], how="inner", on="query", suffixes=["_old", "_new"])
eval_df
```

```python
for i, row in eval_df.iloc[25:50].iterrows():
    print(f"Query:\n{row.query}\n")
    print(f"Old response:\n{row.response_old}\n")
    print(f"New response:\n{row.response_new}\n")
    print("\n################################\n")
```

```python
denied_queries = eval_df[eval_df.response_new.str.contains("Sorry, the system was unable to produce a faithful response")]["query"]

for query in denied_queries:
    print(query)
    print("\n")
```
